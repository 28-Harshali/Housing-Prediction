```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
sns.scatterplot(x=df['Annual Income (k$)'],y=df['Spending Score (1-100)'])
plt.show()
```

```python
scan = df[['Annual Income (k$)','Spending Score (1-100)']]
scan.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

# Assuming 'scan' is your dataset
X = StandardScaler().fit_transform(scan)
dbscan = DBSCAN(eps=0.5, min_samples=2)  # Change eps to a positive value
clusters = dbscan.fit_predict(X)
df['Cluster'] = clusters
print(df)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], c=df['Cluster'], cmap='rainbow', marker='o')
plt.title('DBSCAN Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
plt.show()
```
